- May return a **suboptimal solution**

# Problem 3 — Toggle-It Puzzle

### Scenario
You are given an **N×N grid** of binary values:

- `0` = OFF
- `1` = ON

Your goal is to transform the grid from an initial state to a goal state
(typically **all zeros**).

### Toggle Rule: Neighborhood Toggle

A move chooses one cell (by index or (row, col)) and flips:

- the chosen cell itself
- its **up** neighbor (if it exists)
- its **down** neighbor (if it exists)
- its **left** neighbor (if it exists)
- its **right** neighbor (if it exists)

Flipping means:

- `0 → 1`
- `1 → 0`

**No diagonals**.  
If a neighbor is outside the grid (edge/corner), ignore it.

### Example: 2×2 Indexing

Use row-major indexing:

0 1
2 3

- Toggle `0` affects cells `{0, 1, 2}`
- Toggle `1` affects cells `{1, 0, 3}`
- Toggle `2` affects cells `{2, 0, 3}`
- Toggle `3` affects cells `{3, 1, 2}`


### State Representation
Represent a state as a binary string of length `N*N`:

For a 2×2 grid:

- `"0011"` represents:

00

11


- Goal `"0000"` represents:


00

00


### Search Problem Components

#### State Space
All binary grids of size N×N (equivalently all strings of length N² over {0,1}).

#### Initial State
Example: `"0011"`

#### Goal State
Example: `"0000"`


#### Actions

At any state, an action consists of choosing one cell index:

action ∈ {0, 1, …, N² − 1}

Therefore, each state has exactly N² possible actions.


#### Transition Model
Applying action `i` flips the chosen cell and its valid neighbors (up/down/left/right),
producing a new grid configuration.

#### Path Cost
Each toggle costs 1
